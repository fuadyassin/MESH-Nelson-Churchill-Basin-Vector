```r
# Load necessary library
library(ggplot2)

# Generate some random data
set.seed(123)  # Setting seed for reproducibility
data <- data.frame(
  x = 1:100,  # X values from 1 to 100
  y = rnorm(100, mean=50, sd=10)  # 100 random Y values with mean 50 and standard deviation 10
)

# Create a scatter plot
ggplot(data, aes(x=x, y=y)) + 
  geom_point() +  # Add points
  ggtitle("Scatter Plot of Random Values") +  # Add title
  xlab("X Values") +  # X-axis label
  ylab("Y Values")  # Y-axis label
# Load the required library
library(ncdf4)
library(RColorBrewer)
library(fields)
library(scales)  # Load the scales package for the alpha function
library(viridis)
```
